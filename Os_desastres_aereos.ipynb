Existem alguns valores indefinidos representados por *** . Iremos identificar melhor essas colunas que contém esses valores e iremos criar uma função para trata-los.



```python
ocorencia_aeronave = ocorencia_aeronave.apply(lambda x: x.replace('***','INDEFINIDO'))
ocorencia_aeronave = ocorencia_aeronave.apply(lambda x: x.replace('****','INDEFINIDO'))
```

```python
'''
Função que tem como objetivo substituir os valos NaN nas colunas númericas
'''
def convert_NAN_numeric(df,array_coluns):
    for i in array_coluns:
        df[i].fillna(0,inplace = True)
        

'''
Função crianda para pegar a moda( valore mais frequentes) dos valores objetc
'''
def convert_NAN_Object(df,array_coluns):
    from scipy.stats import mode
    for i in array_coluns:
        array= mode(df[i])
        df[i].fillna(array.mode[0], inplace=True)
        array_delete=list(array)
        del(array_delete)

    
    
```

```python
convert_NAN_numeric(ocorencia_aeronave,['relatorio_publicado','saida_pista','quantidade_motores','quantidade_assentos',
                                       'ano_fabricacao','quantidade_fatalidades'])
convert_NAN_Object(ocorencia_aeronave,['aerodromo','status_investigacao','numero_relatorio','dia_publicacao'])
```

### Excluindo colunas desnecessárias

Em nosso DataFrame , temos uma series de colunas não fazem sentido para a nossa análise exploratória . 
Iremos criar novas colunas mais precisas com base das colunas que iremos dropar.

as colunas são :

 - relatorio_publicado

 - dia_extracao
 - sera_investigada

Usaremos os dados da coluna sera_investigada e status_investigacao para criar um status que endica que a investigação se encontra aberta ate os dias de hoje. em Seguida iremos dropar essas duas colunas. 

Sobre a coluna horário, iremos criar uma função que caracterize o turno da ocorrencia (dia , tarde , noite)





```python
def classificacao_turno(tempo):
    horas= tempo[:2]
    hora=int(horas)
    
    if(hora==0 and hora <1):

        return"Meia-Noite"

    elif(hora>=1 and hora<=11):
        return "MANHA"

    elif(hora>=12 and hora<13):
        return "MAIO-DIA"

    elif(hora>=13 and hora<=18):
         return "TARDE"
    else:
         return "NOITE"
    
```

```python

#Criando coluna TURNO de ocorrências.
ocorencia_aeronave['turno']=ocorencia_aeronave['horario'].apply(classificacao_turno)
```

Iremos dropar as colunas que não são nescessárias para a nossa fase de análise exploratória.

```python

ocorencia_aeronave.drop(['numero_relatorio','sera_investigada','comando_investigador','relatorio_publicado',
          'dia_publicacao','dia_extracao_y','horario','categoria_registro','origem_voo','destino_voo','pais',
                        'codigo_operador','dia_extracao_x','codigo_aeronave','matricula','quantidade_recomendacoes','peso_maximo_decolagem'], axis=1 , inplace=True)
```

# Análise exploratórias

Iremos centrar as ocorrências somentes registradas no território brasileiro. 

```python
ocorrencia_arerias_brasileira =ocorencia_aeronave.loc[ocorencia_aeronave['pais_registro']=="BRASIL"]
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

### Insight 01 Tipos de aeronaves com mais fatalidades

```python
ocorrencia_arerias_brasileira.groupby('tipo_motor')['quantidade_fatalidades'].sum().sort_values().plot(kind='barh' ,color='blue')
```

No grafico acima mostra que as aeronaves com pistão tendem a ocorrer mais fataliades. Seguindo os jatos com mais incidentes. 


### Insight 02 Quais são as maiores causas de acidentes

```python
ocorrencia_arerias_brasileira.groupby('tipo')['quantidade_fatalidades'].sum().sort_values().plot(kind='barh' ,color='blue', figsize=(10,10))


plt.title('QUANTIDADE DE OCORRÊNCIAS AGRUPADAS POR TIPO \n Total: ' + str(ocorrencia_arerias_brasileira['quantidade_fatalidades'].sum()) + ' ocorrências')
plt.ylabel('TIPO DE OCORRÊNCIA')
plt.xlabel('QUANTIDADE DE OCORRÊNCIA')
plt.show()

```

No gráfico acima mostra uma listagem com os tipo de acidentes.
As maiores causas dos acidentes aéreos são por perda de controle em voo. Seguido por perda de controle em solo


### Insight 03 Quais categorias das areronaveis tem mais acidentes

```python
ocorrencia_arerias_brasileira.groupby('categoria_aviacao')['classificacao'].value_counts().sort_values().plot(kind='barh' ,color='blue', figsize=(10,10))

plt.title('QUANTIDADE DE OCORRÊNCIAS AGRUPADAS POR TIPO \n Total: ' + str(ocorrencia_arerias_brasileira['quantidade_fatalidades'].sum()) + ' ocorrências')
plt.ylabel('TIPO DE OCORRÊNCIA')
plt.xlabel('QUANTIDADE DE OCORRÊNCIA')
plt.show()
```

 Nesse gráfico mostra uma listagem dos grupos de categorias e seus respectivos acidentes liderados por aidentes em voos particulares

### Quais tipos de aeronaves ocorrem mais ocorrências

```python
ocorrencia_arerias_brasileira['equipamento'].value_counts().sort_values().plot(kind='bar', color="blue",figsize=(11,5))
```

As maiores ocorrências areas são aviões por ser o equipamento aereo mais utilizado

### Quais fabricantes tiveram mais ocorrências 

```python
ocorrencias_fabricante = ocorrencia_arerias_brasileira.groupby('fabricante')['quantidade_fatalidades'].sum().sort_values(ascending = False)
ocorrencias_fabricante.head(10).plot(kind='barh', figsize=(10,5), color="blue")

plt.title('QUANTIDADE DE OCORRÊNCIAS POR FABRICANTE')
plt.xlabel('QUANTIDADE DE OCORRÊNCIAS')
plt.show()
```

Olha nossa empresa brasileira ai ocupando a 6º posição :D

### Como estão as ocorrências classificadas pelos tipos de dados 

```python
ocorrencia_arerias_brasileira.groupby('fase_operacao')['quantidade_fatalidades'].sum().sort_values().plot(kind='barh' , color='blue', figsize=(10,10))
```

Nesse gráfico mostra que a maioria das ocorrência são originadas na fase de cruzeiro seguido pela fase de pouso

### Nos anos que passaram, os acidentes aumentaram ou diminuíram

```python
data = ocorrencia_arerias_brasileira.copy()
data['dia_ocorrencia'] = data['dia_ocorrencia'].dt.year
data.groupby('dia_ocorrencia')['quantidade_fatalidades'].sum().plot(color='purple', figsize=(12,5))

plt.title('QUANTIDADE DE OCORRÊNCIA AO LONGO DOS ANOS')
plt.xlabel('ANO')
plt.ylabel('QUANTIDADE DE OCORRÊNCIAS')
plt.show()
```
